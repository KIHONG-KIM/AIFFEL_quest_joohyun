# 은행 계좌 만들기

```python
# 은행 계좌 만들기
from math import ceil
from random import randint

# 은행 계좌 클래스 정의
class Account:

  # 클래스 속성(클래스 변수)
  bank_name = 'SC은행'
  total_account_nums = 0    # Account 객체 생성될 때마다 누적되는 값
  total_account_list = []   # Account 객체 생성될 때마다 해당 객체를 리스트에 추가


  # 계좌번호 생성
  # 클래스 속성에 접근하는 메서드여서 클래스 메서드로 정의해 봄. (물론, 일반 인스턴스 메서드로 정의해도 됨)
  @classmethod
  def generate_account_number(cls):
  # def generate_account_number(self):
    while True:
      # 첫 번째 3자리 숫자 생성
      first = str(randint(100, 999))
      # 두 번째 2자리 숫자 생성
      second = str(randint(10, 99))
      # 마지막 6자리 숫자 생성
      last = str(randint(100000, 999999))
      # 계좌 번호 형식으로 조합
      account_number = f"{first}-{second}-{last}"

      # 기존에 생성된 계좌번호 중 중복된 계좌번호가 있는지 체크
      duplicate_list = [acc for acc in cls.total_account_list if acc.account_no == account_number]
      if len(duplicate_list) == 0:    # 중복이 없으면
        break
      else:   # 중복이 있으면 다시 계좌번호 생성
        continue

    return account_number


  # 지금까지 생성된 총 계좌수 리턴
  # 클래스 속성에 접근하는 메서드여서 클래스 메서드로 정의해 봄. (물론, 일반 인스턴스 메서드로 정의해도 됨)
  @classmethod
  def get_account_num(cls):
  # def get_account_num(self):
    return cls.total_account_nums
    # return Account.total_acc_nums   # 일반적인 인스턴스 메서드에서 클래스 속성에 접근할 때는 '클래스명.클래스속성명' 형태로 접근.
    # return self.total_acc_nums    # 또한, 일반적인 인스턴스 메서드에서는 self를 통해 클래스 속성에 접근하는 것도 가능! (단, 읽기만 가능. 수정할 때는 무조건 '클래스명.클래스속성명' 형태 또는 클래스 메서드에서 'cls.클래스속성' 형태로만 가능)


  # 지금까지 생성된 전체 계좌에서 특정 금액 이상의 계좌 정보만 리스트 형태로 리턴
  # 클래스 속성에 접근하는 메서드여서 클래스 메서드로 정의해 봄. (물론, 일반 인스턴스 메서드로 정의해도 됨)
  @classmethod
  def search_account_by_amount(cls, amount):
  # def get_account_num(self, amount):
    # amount가 정수 타입인지 체크
    if not isinstance(amount, int):
      raise ValueError(f"검색에 사용되는 금액 조건이 정수 형식이 아니어서 검색에 실패하였습니다. 다시 검색해 주세요.\n금액 조건: {amount}")

    # 조건에 부합하는 객체들로만 새로운 리스트 생성(리스트 컴프리헨션 사용)
    result = [acc for acc in cls.total_account_list if acc.balance >= amount]
    return result


  def __init__(self, name, amount=0):
    # amount가 정수 타입인지 체크
    if not isinstance(amount, int):
      raise ValueError(f"입금액이 정수 형식이 아니어서 계좌 생성에 실패하였습니다. 다시 계좌를 생성해 주세요.\n입금액: {amount}")

    if amount < 0:
      raise ValueError(f"입금액이 음수로 입력되었습니다.\n입금액: {amount}")

    # self.bank_name = "SC은행"   # 은행명은 클래스 속성으로 수정
    self.name = name
    self.balance = 0    # 잔고 초기화
    self.deposit_no = 0   # 입금 횟수 누적 카운트용 속성 초기화
    self.history_no = 0   # 입출금 내역 고유 번호 초기화
    self.deposit_history_list = []    # 입금 내역 초기화
    self.withdraw_history_list = []   # 출금 내역 초기화

    # 계좌번호는 중복되지 않게 random하게 생성. 형식은 3자리-2자리-6자리
    self.account_no = Account.generate_account_number()

    if amount > 0:
      # amount > 0 이므로 try ~ except 불필요
      # 입금 메서드 호출
      self.deposit(amount)

    # 전체 계좌수 증가
    Account.total_account_nums += 1

    # 전체 계좌 객체를 저장하고 있는 total_acc_list에 추가
    Account.total_account_list.append(self)

    # 계좌 생성 성공 메세지 출력
    print(f"{self.name}님의 계좌가 생성되었습니다. 최초 입금액: {amount:,}원")


  # 입금
  def deposit(self, amount):
    # amount가 정수 타입인지 체크
    if not isinstance(amount, int):
      raise ValueError(f"입금액이 정수 형식이 아니어서 입금 처리를 할 수 없습니다. 다시 입금해 주세요.\n입금액: {amount}")

    # 최소 1원 이상만 가능
    if amount <= 0:
      raise ValueError(f"입금액은 1원 이상이어야 합니다.\n입금액: {amount}")

    # 잔액 업데이트
    self.balance += amount

    # 입출금 내역 번호 증가
    self.history_no += 1

    # 입금 내역에 해당 내용 추가
    history_str = f"{self.history_no}. {amount:,}원 입금, 잔액: {self.balance:,}원"
    self.deposit_history_list.append(history_str)

    # 정상 입금 메세지 출력
    print(f"{self.name} 계좌에 {amount:,}원이 입금되었습니다. 입금 후 잔액: {self.balance:,}원")

    # 입금 횟수 증가
    self.deposit_no += 1

    # 입금 횟수가 5회 누적될 때마다 1% 이자를 잔고에 추가
    if self.deposit_no == 5:
      # 이자 계산
      interest = ceil(self.balance * 0.01)   # 소수점 이하는 무조건 올림으로 처리해서 이자 지급

      # 잔액 업데이트
      self.balance += interest

      # 입금 횟수 다시 초기화
      self.deposit_no = 0

      # 이자 추가 내역도 입금 내역에 추가. 단, 실제 입금은 아니기 때문에 입금 횟수는 증가시키지 않음.
      # 입출금 내역 번호 증가
      self.history_no += 1

      # 입금 내역에 해당 내용 추가
      history_str = f"{self.history_no}. 이자 {interest:,}원 입금, 잔액: {self.balance:,}원"
      self.deposit_history_list.append(history_str)

      # 이자 입금 메세지 출력
      print(f"{self.name} 계좌에 {interest:,}원의 이자가 입금되었습니다. 입금 후 잔액: {self.balance:,}원")


  # 출금
  def withdraw(self, amount):
    # amount가 정수 타입인지 체크
    if not isinstance(amount, int):
      raise ValueError(f"출금액이 정수 형식이 아니어서 출금 처리를 할 수 없습니다. 다시 출금해 주세요.\n출금액: {amount}")

    # 계좌의 잔고 이상으로 출금 불가
    if amount > self.balance:
      raise ValueError(f'현재 잔액을 초과해서 출금할 수 없습니다.\n현재잔액: {self.balance:,}원, 출금액: {amount:,}원')

    # 잔액 업데이트
    self.balance -= amount

    # 입출금 내역 번호 증가
    self.history_no += 1

    # 출금 내역에 해당 내용 추가
    history_str = f"{self.history_no}. {amount:,}원 출금, 잔액: {self.balance:,}원"
    self.withdraw_history_list.append(history_str)

    # 정상 출금 메세지 출력
    print(f"{self.name} 계좌에서 {amount:,}원이 출금되었습니다. 출금 후 잔액: {self.balance:,}원")


  # 계좌 정보 출력
  def display_info(self):
    # [출력 예시] 은행이름: SC은행, 예금주: 파이썬, 계좌번호: 123-12-123456, 잔고: 10,000원
    print(f"은행이름: {Account.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_no}, 잔고: {self.balance:,}원")


  # 입금 내역 출력
  def deposit_history(self):
    if len(self.deposit_history_list) == 0:
      print("입금 내역이 없습니다.")
    else:
      for history in self.deposit_history_list:
        print(history)


  # 출금 내역 출력
  def withdraw_history(self):
    if len(self.withdraw_history_list) == 0:
      print("출금 내역이 없습니다.")
    else:
      for history in self.withdraw_history_list:
        print(history)
```

```python
# 아이펠 학업 관리 프로그램 만들기
from math import ceil
from random import randint

# 아이펠 클래스 정의
class Aiffel:

  # 클래스 속성(클래스 변수)
  total_grew_nums = 0    # Aiffel 객체 생성될 때마다 누적되는 값
  total_grew_dict = {}   # Aiffel 객체 생성될 때마다 해당 객체를 딕셔너리에 추가 (Key : Aiffel.name)
  sorted_key_list = []    # 객체 원본이 저장된 딕셔너리는 그대로 두고, 객체의 key값으로만 이루어진 별도의 리스트를 생성한 다음 정렬이 필요할 때는 sort 메서드를 사용해서 이 리스트에서만 제자리 정렬을 해서 사용하기 위한 속성.


  @classmethod
  def generate_grew_no(cls):
    # 전체 그루 딕셔너리로부터 key값으로만 이루어진 별도의 리스트를 생성한 다음
    # sort 메서드를 사용해 key값순으로(즉, 객체의 이름순으로) 정렬 후 클래스 속성으로 저장(초기화)
    # (원본 딕셔너리는 정렬되지 않고 이 리스트에서만 정렬되는 것임)
    # 정렬할 때는 sorted 함수 보다는 별도의 메모리 할당 없이 '제자리 정렬'이 가능한 sort 메서드 사용해서 정렬
    # sorted 함수 사용하면 사용할 때마다 새로운 리스트를 생성해서 리턴하기 때문에 매번 메모리 할당이 새로 이루어짐.
    # 물론, 가비지 컬렉터가 동작하기 때문에 대부분의 경우에는 sorted 함수를 사용해도 괜찮지만,
    # 객체를 리스트가 아닌 딕셔너리에 저장하면서 객체의 값도 자주 수정되고 정렬도 빈번하게 발생하는 상황이면
    # 이와 같이 key값으로만 저장된 별도의 리스트를 생성해 운영하는 것이 훨씬 더 메모리 효율적이면서 유연함.
    cls.sorted_key_list = [key for key in cls.total_grew_dict.keys()]
    cls.sorted_key_list.sort()
    # [참고] sort 메서드는 리스트 자체를 제자리 정렬하고, 리턴값은 None임. 따라서, sort 메서드의 리턴값을 사용하면 안됨. 그냥 sort 메서드 호출하는 것으로 끝!
    # 즉, 다음 라인처럼 하면 결국 정렬은 정렬대로 잘 하고 리턴하는 None 값을 다시 cls.sorted_key_list 속성에 저장하는 것이 됨.
    # cls.sorted_key_list = cls.sorted_key_list.sort()

    # [참고] 딕셔너리가 아니라 리스트에 객체를 저장하는 경우에는 다음 라인처럼 그냥 sort 메서드 사용해 제자리 정렬하면 됨.
    # 하지만, 이 경우에는 리스트 보다는 딕셔너리에 저장하는 것이 remove 메서드 사용할 때도 특정 요소에 접근하기가 더 용이하기 때문에
    # 리스트 보다는 딕셔너리를 사용하는 것임.
    # cls.total_grew_list.sort(key=lambda g: g.name)

    for i, key in enumerate(cls.sorted_key_list):
      grew = cls.total_grew_dict[key]
      if grew.batch_no == '온라인8기' and grew.course == '코어':
        grew_no = f"{grew.year[2:]}" + '31' + '2' + str(i+1).zfill(2)
        grew.no = grew_no
      else:
        grew.no = "YYNNCXX"


  @classmethod
  def display_all_grews_info(cls):
    for key, grew in cls.total_grew_dict.items():
      print(f"{key}: " + grew.display_info())
      # print(f"{key}: " + repr(grew))


  # classmethod
  # def sort_by(cls, key, reverse=False):
  #   if key == 'name':
  #     cls.sorted_key_list = sorted([key for key in cls.total_grew_dict.keys()], reverse=reverse)
  #     cls.sorted_key_list.sort(reverse)


  @classmethod
  def remove_grew(cls, grew):
    pass


  @classmethod
  def make_groups(cls):
    pass


  def __init__(self, name, year, batch_no, course):
    self.name = name
    self.year = str(year)
    self.batch_no = batch_no
    self.course = course

    # 학번은 모든 그루가 생성된 후에 generate_grew_no 메서드 통해서 한꺼번에 부여함.

    self.point = 0    # 퀘스트 점수
    self.penalty = 0    # 페널티 점수

    # 전체 그루수 증가
    Aiffel.total_grew_nums += 1

    # 전체 그루 객체를 저장하고 있는 total_grew_list에 추가
    Aiffel.total_grew_dict[self.name] = self

    # 그루 생성 성공 메세지 출력
    print(f"{self.name} 객체가 생성되었습니다.")


  # def __repr__(self):
  #   return f"이름: {self.name}, 학번: {self.no}, 점수: {self.point}, 페널티: {self.penalty}, 총점: {self.point - self.penalty}"


  # 그루 정보 출력
  def display_info(self):
    return f"이름: {self.name}, 학번: {self.no}, 점수: {self.point}, 페널티: {self.penalty}, 총점: {self.point - self.penalty}"
    # print(repr(self))


  def add_point(self, point):
    pass


  def add_penalty(self, penalty):
    pass


  def display_scores(self):
    pass
```
